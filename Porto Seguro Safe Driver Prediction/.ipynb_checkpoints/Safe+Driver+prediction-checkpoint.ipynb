```python
#task: you will predict the probability that an auto insurance policy holder files a claim.
#In the train and test data, features that belong to similar groupings are tagged as such in 
#the feature names (e.g., ind, reg, car, calc). In addition, feature names include the postfix 
#bin to indicate binary features and cat to indicate categorical features. Features without 
#these designations are either continuous or ordinal. Values of -1 indicate that the feature 
#was missing from the observation. The target columns signifies whether or not a claim was filed for that policy holder.
#train.csv contains the training data, where each row corresponds to a policy holder, and the target columns 
#signifies that a claim was filed.
import pandas as pd
import numpy as np
import statistics as st
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\Sayesha\Desktop\Machine Learning\Projects\Porto Seguro Safe Driver Prediction\train.csv")
print(data.shape)
#print(data.dtypes)
X = data.drop(["id","target"],axis=1)
y = data["target"]
labels = data["target"].value_counts().index
plt.pie(data["target"].value_counts().values,labels = labels,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
print(type(X))
#imbalanced data
```

```python
#imputation of missing values(-1)
bin_col=[]
cat_col=[]
normal_col=[]
n = 30
print("Number of columns in X:" + str(X.shape[1]))
for col in X.columns:
    z = X.loc[X[col]== -1]
    print(str(col) + ":"+ str(z.shape[0]))

print("Columns with missing values greater than"+ str(n) +"%:") 
for col in X.columns:
    z = X.loc[X[col]== -1]
    if z.shape[0] > (n/100)*(X.shape[0]):
        print(col)
        X = X.drop([col],axis=1)
print("Number of columns in X(after dropping):" + str(X.shape[1]))    
print(type(X))     
        
```

```python
for col in X.columns :
    if 'bin' in col:
        new = X[X[col]!= -1]
        X.loc[X[col]== -1,col] = new[col].value_counts().index[0]
        bin_col.append(col)
        
    elif 'cat' in col:
        new = X[X[col]!= -1]
        X.loc[X[col]== -1,col] = new[col].value_counts().index[0]
        cat_col.append(col)
    
    else:
        new = X[X[col]!= -1]
        X.loc[X[col]== -1,col] = new[col].mean()
        normal_col.append(col)

print(X[normal_col].describe())
print(type(X))
```

```python
#continuous columns
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter

zeroes = data[data["target"]==0]
ones = data[data["target"]==1]

#how to plot subplots ?
#in percentage
for col in normal_col:
    print([col])
    #plt.subplot(2,1,1)
    plt.hist(zeroes[col], bins=50,weights=np.ones(len(zeroes[col])) / len(zeroes[col]))
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    plt.show()  
    #plt.subplot(2,2,1)
    plt.hist(ones[col], bins=50,weights=np.ones(len(ones[col])) / len(ones[col]))
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    plt.show()    
```
