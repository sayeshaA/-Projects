```python
#Encoding categorical data
#OnehotEncoder
X = Benz.iloc[:,2:379]
X_t = test.iloc[:,1:378]
Y_t = test.iloc[:,1]
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
for a in range(0,8):
    X.iloc[:,a] = labelencoder.fit_transform(X.iloc[:,a])
#print(X) 

for a in range(0,8):
    X_t.iloc[:,a] = labelencoder.fit_transform(X_t.iloc[:,a])
#print(X_test) 

from sklearn.decomposition import FastICA, PCA
ica = FastICA(n_components = 20)
a = ica.fit_transform(X)
b = ica.mixing_
print(a.shape)

#train and test set 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(a, Y, test_size=0.33, random_state=42)
```

```python
#ICA and PCA
```

```python
#linear regression 
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(X_train,y_train)
predictions = model.predict(X_test)
print(predictions.shape)
print("Coeff of determination (R2):" + str(model.score(X_train,y_train)))     

from sklearn import metrics
print("RMSE(linear Regression)")
print(np.sqrt(metrics.mean_squared_error(y_test,predictions)))

list1=[]
for a in range(0,1389):
     list1.append(a)
#print(len(list1))

#Decision Tree
from sklearn.decomposition import FastICA, PCA
for b in range(10,20):
        ica = FastICA(n_components = b, max_iter=200)
        a = ica.fit_transform(X)
        b = ica.mixing_

#train and test set 
        from sklearn.model_selection import train_test_split
        X_train, X_test, y_train, y_test = train_test_split(a, Y, test_size=0.33, random_state=42)
        import numpy as np
        from sklearn.tree import DecisionTreeRegressor
        import matplotlib.pyplot as plt

        list1 = []
        list2 = []
        for a in range(2,30):
                list1.append(a)
                regr_3 = DecisionTreeRegressor(max_depth=a)
                regr_3.fit(X_train, y_train)
                y_3 = regr_3.predict(X_test)
                z = np.sqrt(metrics.mean_squared_error(y_3,predictions))
                list2.append(z)
        mean_squared_error = pd.DataFrame({"Numbers": list1,"MSE":list2})    
        #print(mean_squared_error)

        plt.figure()
        ax = sns.scatterplot(x="Numbers",y="MSE",data = mean_squared_error)
        plt.show()

#Fit regression model
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=10)
regr_1.fit(X_train, y_train)
regr_2.fit(X_train, y_train)

#Predict
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)

print("Decision Tree(Regression)")
print(np.sqrt(metrics.mean_squared_error(y_1,predictions)))
print(np.sqrt(metrics.mean_squared_error(y_2,predictions)))
```
